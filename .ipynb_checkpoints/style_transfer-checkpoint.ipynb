```python
import cv2
import torch

import numpy as np
import torch.optim as optim
import matplotlib.pyplot as plt

from copy import deepcopy
from torchvision import models, transforms
from tqdm import tqdm
```

```python
device = ("cuda" if torch.cuda.is_available() else "cpu")
```

```python
lr = 1
alpha = 1e-7
beta = 1e-5

input_image_path = "images/input1.jpg"
style_image_path = "images/style4.jpg"
```

```python
def show(image):
    fixed_color_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    fixed_color_img = np.array(fixed_color_img / 255).clip(0, 1)

    plt.imshow(fixed_color_img)
    plt.show()
    
def init(tensor):
    _, C, H, W = tensor.shape
    
    return torch.randn(C, H, W).mul(0.001).unsqueeze(0)
```

```python
def style_loss(g, s, loss_function):
    C = g.shape[0]
    
    return loss_function(g, s) / (C**2)
```

```python
def im2tensor(image):
    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.ToTensor()
    ])
    
    normalize = transforms.Normalize([103.939, 116.779, 123.68], [1,1,1])
    
    return normalize(transform(image) * 255).unsqueeze(dim=0)

def tensor2im(tensor):
    transform = transforms.Compose([
        transforms.Normalize([-103.939, -116.779, -123.68],[1,1,1])
    ])
    
    tensor = tensor.squeeze()

    return transform(tensor).cpu().numpy().transpose(1, 2, 0)
```

```python
input_image = cv2.imread(input_image_path)
style_image = cv2.imread(style_image_path)

show(input_image)
show(style_image)
```

```python
vgg = models.vgg19(pretrained=True)

model = deepcopy(vgg.features)
model.to(device)

for p in model.parameters():
    p.requires_grad = False
```

```python
input_tensor = im2tensor(input_image).to(device)
style_tensor = im2tensor(style_image).to(device)

g = init(input_tensor).to(device).requires_grad_(True)
```

```python
def forward(tensor):
    activation_layers = {
        '3': 'relu1_2',
        '8': 'relu2_2',
        '17': 'relu3_3',
        '26': 'relu4_3',
        '35': 'relu5_3',
        '24': 'relu4_2'
    }
    
    features = {}
    x = tensor

    for name, layer in model._modules.items():
        x = layer(x)

        if name in activation_layers:
            if name == '31' or name == '24':
                features[activation_layers[name]] = x
            else:
                _, C, H, W = x.shape

                y = x.view(C, H * W)
                y = torch.mm(y, y.t())
                features[activation_layers[name]] = y / (H * W)
            
            if name == '35':
                break

    return features
```

```python
def transfer_style(optimizer, loss_function, amount=10000):
    input_layers = ['relu4_2']
    input_weights = {'relu4_2': 1}

    style_layers = ['relu1_2', 'relu2_2', 'relu3_3', 'relu4_3', 'relu5_3']
    style_weights = {'relu1_2': 0.2, 'relu2_2': 0.2, 'relu3_3': 0.2, 'relu4_3': 0.2, 'relu5_3': 0.2}
    
    input = forward(input_tensor)
    style = forward(style_tensor)
    
    def closure():
        optimizer.zero_grad()
        
        next = forward(g)

        loss_input = 0
        loss_style = 0

        for j in input_layers:
            loss_input += input_weights[j] * loss_function(next[j], input[j])
        for j in style_layers:
            loss_style += style_weights[j] * style_loss(next[j], style[j], loss_function)

        total_loss = alpha * loss_input + beta * loss_style
        total_loss.backward(retain_graph=True)

        return total_loss
    
    for _ in tqdm(range(amount)):
        optimizer.step(closure)
```

```python
loss_function = torch.nn.MSELoss()
optimizer = optim.Adam([g], lr=lr)

transfer_style(optimizer, loss_function, 50000)

show(tensor2im(g.detach().clone()))
```
